# Example 1 (age and sex)
This small example dataset contains only two attributes: ```age``` (numeric) and ```sex``` (categorical).

````python
# Load dataset and show the distributions
import DART

## set constants
random_state = 1337

data = DART.toy.age_and_sex(random_state=random_state)
DART.toy.show_distributions(data);
````

````python
# create the DART Dataset and compare age distributions of different sexes
import pprint

dataset = DART.Dataset(data)
print(dataset)
comparison = dataset.compare(
    {"sex": "Female"},
    {"sex": "Male"}
).export("maximal")
print(pprint.pformat(comparison))
````
